```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px

plt.style.use('fivethirtyeight')
pd.set_option('display.max_columns', 500)
color_pal = plt.rcParams['axes.prop_cycle'].by_key()['color']
from fredapi import fred
```

```python
from google.colab import userdata
fred_key = userdata.get("FRED_API")
```

#1. Create the FRED Object

```python
fred_obj = fred.Fred(api_key= fred_key)
```

# 2. Search for Economic Data

```python
sp_search = fred_obj.search('S&P', order_by= 'popularity')
sp_search.shape #Limit of 1000 rowswith search default value
sp_search
```

# 3. Pull and Plot Raw Data

```python
sp500 = fred_obj.get_series(series_id= 'SP500')   #Returns a pandas series with index as datetime column and values as price of SP500
```

```python
px.line(sp500, labels = {'index':'Year', 'value':'Unemployment Rate'}).update_layout(showlegend = False)
```

```python
unemployment_df = fred_obj.search(
    'Unemployment rate state')
```

```python
filtered_df = unemployment_df[
    (unemployment_df['frequency'] == 'Monthly') & (unemployment_df['seasonal_adjustment'] == "Seasonally Adjusted") & (unemployment_df['units'] == 'Percent')
    ]
filtered_df
```

```python
test1 = filtered_df
test1.shape
```

```python
test1 = test1[test1['title'].str.contains('Unemployment Rate in')]
test1.shape
```

```python
id_to_state = test1['title'].str.replace('Unemployment Rate in', '').to_dict()
```

```python
filtered_df.shape
```

```python
#filtered_df['title'].str.contains('Unemployment Rate')
test1[test1['title'].str.contains('Unemployment Rate')].shape
```

```python
test1.index
```

```python
results = []
for id in test1.index:
  result = fred_obj.get_series(id).to_frame(name = id)
  #break   #Inspect script and check first id if working properly
  results.append(result)
results
```

```python
type(results)
```

```python
results_df = pd.concat(results, axis = 1) #axis = 1 stack side by side
unemployment_states = results_df.drop(results_df.columns[0], axis = 1)

unemployment_states  #NaN values for some ids which may be due to inavailability
```

```python
px.line(unemployment_states.isna().sum(axis = 1)) #Used this to identify a threshold of sum of missing values to be dropped
#issing values for each date
#For checking missing values which are more apparent before 1980 where monthly updates were not yet available
```

```python
cleaned_unemp_states = unemployment_states.dropna(axis = 0)
cleaned_unemp_states.columns = [id_to_state[state] for state in cleaned_unemp_states.columns]
```

```python
#Plot States Unemployment Rate
px.line(cleaned_unemp_states , labels = {'value':'Unemployment Rate', 'index':'Year'}).update_layout(legend_title_text = 'State', title = {'text':'States Unemployment Rate',
                                                                                                                                           'x':0.5, 'xanchor':'center'})

#Double-click on each state or multiple states to  isolate them
```
